```python
import json
import random

def generate_oriented_graph(size, min_out_degree):
    """
    Generate a random oriented graph with a given size and a minimum out degree.
    """
    if size <= 0 or min_out_degree < 0 or min_out_degree >= size:
        raise ValueError("Invalid parameters: size must be > 0, min_out_degree must be >= 0 and < size.")
        
    graph = []
    for i in range(size):
        graph.append([0] * size) #initializd all edges to - (no edges)
        
    for i in range(size):
        available_nodes = []
        for j in range(size):
            if i != j and graph[i][j] == 0 and graph[i][j] == 0:
                available_nodes.append(j)
                
        for _ in range(min_out_degree):
            if not available_nodes:
                break;
            target = random.choice(available_nodes)
            graph[i][target] = 1 #add directed edge from 1 to target
            available_nodes.remove(target)
            
    return(graph)
        
def generate_unique_graphs(num_graphs, size, num_edges):
    """
    Generate a list of unique oriented graphs.
    """
    unique_graphs = set()
    while len(unique_graphs) < num_graphs:
        graph = generate_oriented_graph(size, num_edges)
        graph_tuple = tuple(tuple(row) for row in graph)
        unique_graphs.add(graph_tuple)

    return [list(graph) for graph in unique_graphs]

def save_graphs_to_json(graphs, filename):
    """
    Save a list of graphs to a JSON file.
    """
    with open(filename, 'w') as file:
        json.dump(graphs, file)

if __name__ == "__main__":
    num_graphs = 10000
    size = 100  # You can adjust the size as needed
    num_edges = 3  # You can adjust the number of edges as needed

    graphs = generate_unique_graphs(num_graphs, size, num_edges)
    save_graphs_to_json(graphs, 'unique_graphs.json')

    print(f"Generated {len(graphs)} unique graphs and saved to 'unique_graphs.json'.")
```

```python
def calculate_total_nodes(min_degree):
    if min_degree < 2:
        return 1  # Minimum degree of 1 requires at least 1 node

    # Initial central node
    total_nodes = 1
    current_node = 0

    while True:
        num_new_nodes = min_degree - 1
        if total_nodes + num_new_nodes > current_node + 1:
            break
        total_nodes += num_new_nodes + 1
        current_node += num_new_nodes + 1

    return total_nodes

# Example usage
min_degree = 3
total_nodes = calculate_total_nodes(min_degree)
print(f"Total number of nodes for minimum degree {min_degree}: {total_nodes}")
```

```python
import random
import json

def build_directed_oriented_graph(min_degree):
    from itertools import combinations
    
    def add_edges(graph, nodes):
        # Connect each pair of nodes in the subset with directed edges
        for i, j in combinations(nodes, 2):
            graph[i][j] = 1  # Forward edge only
    
    size = min_degree * (min_degree - 1) // 2 + min_degree  # Estimate the size of the graph
    graph = [[0] * size for _ in range(size)]
    
    current_node = 0
    all_nodes = []  #Track all nodes that have been added so far
    
    while current_node < size:
        # Determine the number of new nodes
        num_new_nodes = min_degree - 1
        
        if current_node + num_new_nodes >= size:
            num_new_nodes = size - current_node - 1
        
        new_nodes = list(range(current_node + 1, current_node + 1 + num_new_nodes))
        
        all_nodes.extend(new_nodes)
        
        # Connect the central node to new nodes
        for node in new_nodes:
            graph[current_node][node] = 1  # Oriented: only forward edge
        
        # Fully connect new nodes to each other with oriented (one-way) directed edges
        add_edges(graph, new_nodes)
        
        #Ensure that the first node is connected to its neighbors
        if current_node == 0:
            for node in new_nodes:
                graph[0][node] = 1 # Oriented edge from the first node to the new nodes
                
        current_node += num_new_nodes + 1
    
    return graph


def build_directed_oriented_graph_with_backward_arcs(min_degree):
    from itertools import combinations
    
    def add_edges_with_backward(graph, nodes):
        # Connect each pair of nodes in the subset with either a forward or backward edge
        for i, j in combinations(nodes, 2):
            if graph[j][i] == 0:  # Ensure there's no backward edge before adding forward edge
                graph[i][j] = 1  # Forward edge
            else:
                graph[i][j] = 0  # Remove forward edge if backward exists (to maintain orientation)
    
    size = min_degree * (min_degree - 1) // 2 + min_degree  # Estimate the size of the graph
    graph = [[0] * size for _ in range(size)]
    
    current_node = 0
    all_nodes = [0] # Track all nodes that have been added so far
    
    while current_node < size:
        # Determine the number of new nodes
        num_new_nodes = min_degree - 1
        
        if current_node + num_new_nodes >= size:
            num_new_nodes = size - current_node - 1
        
        new_nodes = list(range(current_node + 1, current_node + 1 + num_new_nodes))
        
        # Add new nodes to the list of all nodes
        all_nodes.extend(new_nodes)
        
        # Connect the central node to new nodes
        for node in new_nodes:
            graph[current_node][node] = 1 #Oriented: Only forward edge
        
        # Fully connect new nodes to each other with directed edges and selective backward arcs
        add_edges_with_backward(graph, new_nodes)
        
        # Ensure that the first node is connected to its neighbors
        if current_node == 0:
            for node in new_nodes:
                graph[0][node] = 1 # Oriented edge from the first node to the new nodes
        
        current_node += num_new_nodes + 1
    
    return graph


def generate_graph_data(num_graphs, min_degree_range, with_backward_arcs):
    graph_data = []
    for _ in range(num_graphs):
        min_degree = random.randint(*min_degree_range)
        if with_backward_arcs:
            graph = build_directed_oriented_graph_with_backward_arcs(min_degree)
        else:
            graph = build_oriented_graph_with_dsp(min_degree)
        
        size = len(graph)
        num_sets = len([1 for i in range(size) if graph[i].count(1) >= min_degree])
        
        graph_data.append({
            'min_degree': min_degree,
            'size': size,
            'num_sets': num_sets,
            'graph': graph
        })
    
    return graph_data

def save_graph_data_to_json(filename, graph_data):
    with open(filename, 'w') as f:
        json.dump(graph_data, f, indent=2)

# Generate 100 graphs without backward arcs
graphs_no_backward = generate_graph_data(100, (3, 25), with_backward_arcs=False)
save_graph_data_to_json('graphs_no_backward.json', graphs_no_backward)

# Generate 100 graphs with backward arcs
graphs_with_backward = generate_graph_data(100, (3, 25), with_backward_arcs=True)
save_graph_data_to_json('graphs_with_backward.json', graphs_with_backward)
```

```python
def build_oriented_graph_with_dsp(min_degree):
    from itertools import combinations
    import random
        
    size = min_degree * (min_degree - 1) // 2 + min_degree # Estimate the size of the graph
    graph = [[0] * size for _ in range(size)]

    def add_edges_with_dsp(node, neighbors):
        """Ensure each node's first-degree neighbors connectivity for DSP """
        for i in neighbors:
            # Connect each first-degree neighbor to at most min_degree second_degree neighbors of node
            possible_second_degree_neighbors = [n for n in range(size) if n != node and n not in neighbors]
            selected_second_neighbors = random.sample(possible_second_degree_neighbors, min(min_degree, len(possible_second_degree_neighbors)))
            for j in selected_second_neighbors:
                graph[i][j] = 1
                
    for current_node in range(size):
        # Determine the number of new nodes
        num_new_nodes = min_degree - 1
        
        if current_node + num_new_nodes >= size:
            num_new_nodes = size - current_node - 1
            
        new_nodes = list(range(current_node + 1, current_node + 1 + num_new_nodes))
        
        # Connect the current node to the new nodes
        for node in new_nodes:
            graph[current_node][node] = 1 # Oriented: only forward edge
            
        # Fully connect new nodes to each other with directed edges
        add_edges_with_dsp(graph, new_nodes)
    return graph
```
