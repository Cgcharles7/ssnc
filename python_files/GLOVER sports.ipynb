```python
import random
import json
from datetime import datetime, timedelta

# Sample dataset for quarterbacks and wide receivers
quarterbacks = [
    "Patrick Mahomes", "Tom Brady", "Josh Allen", "Aaron Rodgers", 
    "Justin Herbert", "Joe Burrow", "Lamar Jackson", "Jalen Hurts"
]

wide_receivers = [
    "Davante Adams", "Tyreek Hill", "Justin Jefferson", "Cooper Kupp", 
    "Stefon Diggs", "Deebo Samuel", "Ja'Marr Chase", "A.J. Brown"
]

# Sample outcomes of a pass
pass_results = [
    "Complete", "Incomplete", "Interception", "Touchdown"
]

# Sample yardage for passes
yardage = list(range(-5, 100))  # -5 for losses to 99 yards

# Function to generate a football passing dataset
def generate_football_dataset(num_entries=50):
    dataset = []
    start_date = datetime(2024, 1, 1)  # Example start date

    for _ in range(num_entries):
        date = start_date + timedelta(days=random.randint(0, 30))  # Random day in January
        quarterback = random.choice(quarterbacks)
        receiver = random.choice(wide_receivers)
        result = random.choice(pass_results)
        yards = random.choice(yardage)
        
        # Add an entry to the dataset
        dataset.append({
            "date": date.strftime("%Y-%m-%d"),
            "quarterback": quarterback,
            "receiver": receiver,
            "result": result,
            "yards": yards
        })

    # Save the dataset as a JSON file
    with open("football_passing_dataset.json", "w") as json_file:
        json.dump(dataset, json_file, indent=4)

    return dataset

# Example usage
if __name__ == "__main__":
    football_data = generate_football_dataset(100)  # Generate 100 rows of data
    for play in football_data[:50]:  # Print the first 5 rows
        print(play)
```
